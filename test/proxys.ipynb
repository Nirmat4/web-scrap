```python
import requests
from concurrent.futures import ThreadPoolExecutor

def test_proxy(proxy):
    try:
        response = requests.get(
            "https://www.autocompara.com",  # Prueba con tu sitio real
            proxies={"http": f"http://{proxy}", "https": f"http://{proxy}"},
            timeout=15
        )
        if response.status_code == 200:
            print(f"✅ {proxy}")
            return proxy
    except Exception as e:
        print(f"❌ {proxy} - Error: {str(e)[:30]}")
        return None

# Descarga proxies de una fuente (ejemplo)
def get_free_proxies():
    url = "https://api.proxyscrape.com/v2/?request=getproxies&protocol=http"
    response = requests.get(url)
    return response.text.splitlines()

# Uso:
if __name__ == "__main__":
    proxies = get_free_proxies()  # O usa proxies_gratuitos de arriba
    valid_proxies = []
    
    with ThreadPoolExecutor(max_workers=20) as executor:
        results = executor.map(test_proxy, proxies)
        valid_proxies = [p for p in results if p]
    
    print(f"\nProxies válidos ({len(valid_proxies)}):")
    print("\n".join(valid_proxies))
```
